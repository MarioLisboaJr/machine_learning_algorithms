### Introdução ao Algoritmo

KNN é um algoritmo de aprendizagem supervisionada onde temos a possibilidade de utilização tanto para problemas de classificação quanto para regressão.
Na classificação a máquina irá dizer a que grupo determinado registro faz parte, já na regressão irá nos fornecer um número/valor para dado fato/acontecimento.

Este algoritmo é do *tipo lazy*, ou, conhecido como de aprendizado lento. Os algoritmos do *tipo lazy*, não necessitam de dados de treinamento para se gerar o modelo, o que diminui em partes o processo inicial, mas em contrapartida gerará uma necessidade de análise posterior mais apurada. No caso de algoritmos que não necessitam de treinamento, todos os dados obtidos no dataset serão utilizados na fase de teste, resultando em um treinamento muito rápido e em um teste e validação lentos, momento o qual necessitamos estar bem atentos aos resultados gerados.

### Como Funciona

Resumindo grosseiramente, ele se baseia em quão similar um dado é do outro.

Tomemos como exemplo a figura abaixo. Possuímos duas classes de dados, bolinhas azuis e bolinhas vermelhas, e uma terceira bolinha verde ainda não classificada, onde gostaríamos de classificá-la em azul ou vermelha.

Intuitivamente classificaríamos esta bolinha verde em vermelha, visto que seus dados 'vizinhos' mais próximos são em sua maioria vermelhos. O funcionamento do algoritmo é exatamente este, você define os k vizinhos mais próximos aos dados a ser analisado para dizer qual classificação do novo dado. Por exemplo, na figura temos k=3.

Como a proximidade dos dados é um fator importante, torna-se necessário seu cálculo em distância. Podendo ela ser:<br/>

-Distância Euclidiana<br/>
-Distância de Hamming<br/>
-Distância Manhattan<br/>
-Distância de Markowski<br/>
-Distância de Minkowski<br/>

E como escolher o melhor valor para k?<br/>

Não existe um número mágico, pois cada dataset tem sua particularidade. De uma forma geral, cientistas testam valores ímpares para k, pois identificou-se resultados melhores com esses números. A melhor maneira de definir o k acaba sendo por meio de testes no modelo.

**Sendo assim, temos que as etapas do algoritmo são:**<br/>

**1) Receber um dado não classificado**<br/>
**2) Definir a distância do novo dado em relação a todos os outros dados já classificados**<br/>
**3) Selecionar os k vizinhos com menores distancias**<br/>
**4) Analisar qual a classe em maioria dentre os k vizinhos**<br/>
**5) Tomar como resultado a classe que mais aparece**<br/>

### Atenção na Utilização do Algoritmo

O algoritmo KNN em geral obtém melhores resultados em uma baixa volumetria de dados.<br/>

Quando o k é pequeno, a classificação fica mais sensível a regiões bem próximas (podendo ocorrer o problema de *overfitting*). Com k grande, a classificação fica menos sujeita a ruídos podendo ser considerada mais robusta, porém se k for grande demais, pode ser que haja o problema de *underfitting*.<br/>

Caso a quantidade de *features* no seu modelo sejam consideravelmente muitas, será necessária uma realimentação do seu modelo também em registros, ou seja, quanto mais colunas tivermos mais dados serão necessários, mas muito cuidado, *overfiting* ocorrem quando estamos em uma matriz esparsa.<br/> 
Lembra que comentamos que a base deste modelo são medidas de distância? Então, em algoritmos que se baseiam em distância para classificar ou predizer valores será necessário um cuidado muito grande quando a escalabilidade de seus dados, ou seja, se tiver uma coluna com valores variando entre 1000 e 1000000 e outra coluna no mesmo dataset que tenha seu espaço entre 0 e 10, as distâncias destes serão muito grandes, gerando um resultado impreciso/incorreto. Para este tipo de algoritmo, indicamos a realização de um pré processo chamando **redução de dimensionalidade** (Dimensionality reduction), o mais conhecido é o PCA (Principal Component Analysis), que na forma mais simples irá alocar a representatividade dos valores das colunas entre um mesmo range, por exemplo entre 0 e 1. Lembrando que os números são infinitos dentro de suas escalas, permitindo assim esta representação dentro de um mesmo plano, desta forma obteremos resultados mais precisos e mais seguros, quando utilizarmos algoritmos que se baseiam em medidas de distância.

### Prós e Contras

**Prós**<br/>
A favor do KNN temos:<br/>
- É um dos mais simples algorítmos a se implementar
- Não necessita de treinamento
- Boa precisão na maioria dos casos aplicados
- Poucos parâmetros existentes (Valor de K e Medida de Distância)
- Pode ser usado tanto para classificação quanto para regressão

**Contras**<br/>
Contra o KNN temos:<br/>
- Algoritmo lento
- Não funciona muito bem com muitas variáveis de entrada
- Um desbalanceamento nas quantidades entre os dados classificados pode gerar um enviesamento para a classificação que possui maior quantidade de dados
- Sensibilidade a outliers

**Importar Bibliotecas**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split 
#from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
```

**Importar DataSet**

```python
df = pd.read_csv('mushrooms_dataset.csv', sep=',')
```

**Explorar o DataSet**

```python
#Visualizando o dataset
display(df.head())
```

```python
#Entendendo os dataset, qtd de linhas, colunas, dados faltantes, tipo dos dados
df.info()
```

```python
#Verificando detalhes estatísticos do dataset
display(df.describe().transpose())
```

```python
#Checar o balanceamento dos dados
#Verificar no dataset quantos dos cogumelos são comestíveis e quantos são venenosos
#Pelo dicionário do dataset temos que na coluna 'class' os dados do tipo:
#e (edible) são os comestíveis, e os p (poisonous) são os venenosos

comestiveis = pd.value_counts(df['class'])[0]
venenosos = pd.value_counts(df['class'])[1]

print('\n Cogumelos Comestíveis: {:,}\n Cogumelos Venenosos: {:,}'.format(comestiveis, venenosos))

#Exibir gráfico
plt.figure(figsize=(5,4))
color1 = sns.color_palette('bright')[0]
color2 = sns.color_palette('bright')[3]
plt.pie([comestiveis, venenosos], labels=['Comestiveis', 'Venenosos'], autopct='%1.0f%%', colors=[color1, color2], shadow=True)
plt.title('Percentual de Cogumelos Comestíveis vs. Venenosos')
plt.show()
```

Como os dados entre as classificações Comestíveis e Venenosos estão bem balanceadas, 52 vs. 48%, evitamos um problema de viés na classificação.

**Separar os Dados**

Separar os dados em x e y, sendo:
- y é o resultado, ou seja, a característica a ser classificada. Variável dependente. 
<br/>No nosso caso queremos classificar em comestível/venenoso e esta informação está presente na coluna class. Sendo esta nosso y
- x são as características dos dados que possibilitam encontrar o y. Variável independente.
<br/>No nosso caso o x serão todas a colunas exceto a class

```python
y = df['class']
x = df.drop('class', axis=1)
```

**Criar Variáveis Dummies**

Converter a variável categórica em variáveis fictícias/indicadoras.

Como nossas características do modelo são strings, precisamos transformas estas informações em numéricas para que computacionalmente elas possam ser interpretadas.

Criando variáveis dummies transformamos as classes de cada característica em novas características, porém, com informações em valores binários. Por exemplo, na coluna veil-type temos p para partial e u para universal. Transformando esta coluna em variáveis dummies teremos um novo dataframe com as colunas veil-type_p e veil-type_u com valores 1 ou 0, indicando se verdadeiro ou falso.

```python
x = pd.get_dummies(x)
display(x.head())
```

**Separar Dados de Treino e Dados de Teste**

Anteriormente dissemos que algoritmos do tipo lazy não necessitam de treinamento. Mas aqui, esta etapa será aplicada apenas pra visualização das métricas de avaliação do modelo

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
#test_size para definir o ponto em que os dados serão particionados
#random_state para para que a chamada da função seja replicável. Caso não seja informado os dados serão sempre diferentes
```

**Reduzir a Dimensionalidade**

Etapa informativa. Para este problema ela torna-se desnecessária.

Será demonstrada para frisar o ponto de atenção destacado anteriormente.

..."é necessário um cuidado muito grande quando a escalabilidade de seus dados, ou seja, se tiver uma coluna com valores variando entre 1000 e 1000000 e outra coluna no mesmo dataset que tenha seu espaço entre 0 e 10, as distâncias destes será muito grande, gerando um resultado impreciso/incorreto"...

```python
#scaler = StandardScaler()  
#scaler.fit(x_treino)

#x_treino = scaler.transform(x_treino)  
#x_teste = scaler.transform(x_teste) 
```

**Criar o modelo KNN**

```python
#Definir inicialmente o valor de vizinhos = 5
classificar_cogumelos = KNeighborsClassifier(n_neighbors=5)  
#O default da função calcula a Distância por Minkowski

#Treinar o modelo, com dados de treinamento
classificar_cogumelos.fit(x_treino, y_treino)  

#Prever os valores de y para os dados de teste (x_teste)
y_pred = classificar_cogumelos.predict(x_teste) 
```

**Avaliando o Modelo**

Matriz de Confusão:


```python
#Comparar os valores previstos de y pelo modelo com os valores y_teste
#Criar matriz de confusão
matriz = ConfusionMatrixDisplay(confusion_matrix(y_teste, y_pred), display_labels=classificar_cogumelos.classes_)
matriz.plot()
plt.title('Matriz de Confusão')
plt.show()
```

```python
#Exibir o relatório de classificação
print('\nRelatório de classificação: \n\n', classification_report(y_teste, y_pred))  

#Calcular a acuracidade do modelo
acuracidade = accuracy_score(y_teste, y_pred)
print('\nAcurácia do modelo: {:1.0%}'.format(acuracidade))
```

Pelas métricas de avaliação do modelo, observamos que ele acertou todas as classificações!

**Testar Modelo com Outros Valores de k**

Apesar do nosso modelo já apresentar um resultado muito bom com k = 5. Para demonstração, criaremos um loop de teste para k com outros valores.

```python
#Como é indicado a utilização de valores ímpares para k, vamos criar uma lista de valores a serem testados
valores_k = list(range(1, 50, 2))

#Criar um loop para testes
k_acuracidade = []

for i in valores_k:
    classificar_cogumelos = KNeighborsClassifier(n_neighbors=i)  
    classificar_cogumelos.fit(x_treino, y_treino)  
    y_pred = classificar_cogumelos.predict(x_teste) 
    acuracidade = accuracy_score(y_teste, y_pred)
    k_acuracidade.append(acuracidade)
```

```python
#Exibir resultado do teste para valores de k
plt.figure(figsize=(15,5))
plt.plot(valores_k, k_acuracidade, marker='o', linestyle='--')
plt.xlabel('Valores para k')
plt.ylabel('Acuracidade do Modelo')
plt.xticks(valores_k, valores_k)
plt.show()
```

Pelo gráfico observamos que para k igual a 1, 3 ou 5 teriamos a mesma acurácia do modelo.

Porém, para fugirmos de um possível overfitting*, fica melhor a utilização do k = 5.

*Overfitting (sobre-ajuste) é um termo usado em estatística para descrever quando um modelo estatístico se ajusta muito bem ao conjunto de dados anteriormente observado, mas se mostra ineficaz para prever novos resultados.

**Quais são então as principais características dos cogumelos para diferenciá-los entre comestíveis ou venenosos?**

Para conseguirmos esta informação, podemos checar a correlação de cada característica do cogumelo com ele ser Comestível ou Venoso. Quanto mais próximo o valor da correlação for de |1| mais forte ela se torna, sendo então um grande indicativo.<br/>
Se o valor da correlação for positivo este par segue uma tendencia parecida (se um aumenta o outro também aumenta), agora se o valor for negativo seguem tendencias contrárias (se um aumenta o outro diminui).

```python
#Criar dataframe para análise da correlação
corr = x
corr['class'] = y

#Tranformar a coluna class em variáveis dummies para possibilitar o cálculo da correlação
corr = pd.get_dummies(corr)

#Calcular a correlação entre todas as informações do dataframe
corr = corr.corr()
```

```python
corr_comestiveis = corr['class_e'].sort_values(ascending=False)
corr_venenosos = corr['class_p'].sort_values(ascending=False)
```

```python
print('\nTop 5 características dos cogumelos comestíveis:\n')
print(corr_comestiveis[1:6])
print('\nTop 5 características dos cogumelos venenosos:\n')
print(corr_venenosos[1:6])
```

```python
#Exibir gráfico da correlação
corr = pd.DataFrame()
corr['comestivel'] = corr_comestiveis
corr['venenosos'] = corr_venenosos
corr = corr[1:-2]

plt.figure(figsize=(15,5))
sns.heatmap(corr, yticklabels=False)
plt.title('\nCorrelação da classe do cogumelo com sua característica\n')
plt.show()
```

Pelo gráfico fica claro que as informação entre as classes são simetricamente opostas. Ou seja, a caractrísta mais fortes dos cogumelos comestíveis são as mais fracas dos cogumelos venenosos.
